We want to know the fequence of dominant phenotypes in the f1 from 2 random parents from the whole pop.

We did a lot of figuring out on the white board, the calculation in the end is rather easy.

```python
# we could also read them out from the file
```

```python
# setting up the vaiables for each allel combination and population side
k = 15
m = 29
n = 24
N = k + m + n
```

```python
# Calculating the frequence P of dominant phenotypes
P = k/N + m/(N*(N-1))*((4*k + 3*m + 2*n - 3)/4) + n/(N*(N-1))*((2*k + m)/2)
print(round(P, 5))
```
