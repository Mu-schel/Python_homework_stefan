```python
import util
```

```python
# this made a weird format out of fasta, all lines are strings now separated by comma
list1 = util.read_file_into_lines("/home/uschi/Python_homework_stefan/misc/rosalind_gc.txt")
```

```python
# here we make use of this weird string format, to get to a list of items (strings) always starting with Rosalind_XXXX
list2 = ",".join(list1) # join all strings in the list were the comma is
list3 = list2.replace("," , "") # replace all comma by emptyspcae
list4 = list3.split(">") # sepatate the complete string at the > into a list
list4.remove(list4[0]) # remove the first object in the list which was an empty string
```

```python
# filling a dictionary with the Rosalind_XXXX IDs as keys and sequences as values
fasta_dict = {}
for j in range(len(list4)):
    fasta_dict[list4[j][0:13]] = list4[j][13:]
# print(fasta_dict)  
```

```python
keys = list(fasta_dict.keys()) #make a list of keys so we can loop through specific indices, I know it's not sexy
```

```python
for ID in range(len(fasta_dict)):
    a = fasta_dict.get(keys[ID])
    count_C = a.count("C")
    count_G = a.count("G")
    G_C_count = ((count_C + count_G)*100)/ len(a)
    GC = round(G_C_count , 6)
    print(keys[ID], "\n", GC)
```
