Initializing challenge

```python
values = [1, -2, 3, 4, 1.0]
smallest, largest = None, None
for v in values:
    if smallest is None and largest is None:
        smallest, largest = v, v
    else:
        smallest = min(smallest, v)
        largest = max(largest, v)
print(smallest, largest)
```

Files and how to open them.

```python
file_path = "some file path"

def read_file_into_lines(file_path):
    lines =  []
    with open(file_path,'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip() #just get rid of new line characters the r is for just the right side
            lines.append(cleaned)
    return lines

# we saved this into the util(ity) file so we can open it from everywhere
```
