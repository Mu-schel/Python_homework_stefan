Calculate expected offspring

-----------------
given a string of six numbers each number at max 20k 

each of the six numbers representing a specific parent couple with specific genotypes

return the number of expected offspring if each couple has two offspring


```python
import util
```

```python
list_of_integers = util.read_file_into_lines("/home/uschi/Python_homework_stefan/misc/rosalind_iev.txt")
```

```python
intiger_string = list_of_integers[0]
intiger_list = intiger_string.split(' ')
intiger_list = [int(i) for i in intiger_list]
intiger_list
```

```python
A = intiger_list[0]
B = intiger_list[1]
C = intiger_list[2]
D = intiger_list[3]
E = intiger_list[4]
F = intiger_list[5]
```

```python
Expected_offspring = (2*1*A) + (2*1*B) + (1*2*C) + (2*0.75*D) + (2*0.5*E) + (2*0*F)
print(Expected_offspring)
```
