2. How to get the reverse compliment.

```python
def rev_comp(DNA):
    rev_DNA = ""
    for char in DNA:
        rev_DNA = char + rev_DNA
        rev_dna = rev_DNA.casefold()
    rev_comp_DNA = ""
    for base in rev_dna:
        if base == "t":
            rev_comp_DNA = rev_comp_DNA + "A" 
        elif base == "g":
            rev_comp_DNA = rev_comp_DNA + "C"
        elif base == "c":
            rev_comp_DNA = rev_comp_DNA + "G"
        else: 
            rev_comp_DNA = rev_comp_DNA + "T"

    return rev_comp_DNA
```

4. Given an ORF and a starting position, how to test if a protein exists.

```python
# orf_seq = ["ATG","GGA","GGA","ATG","TAG","GCA","ATG","GGA","GGA","ATG","TAG"]
```

```python
def find_START(orf_seq):
    start = "ATG"
    list_START = []

    for i in range(len(orf_seq)):
        if orf_seq[i] == start:
            list_START.append(i)
    return list_START
```

```python
def find_STOP(orf_seq):
    list_STOP = []

    for i in range(len(orf_seq)):
        if orf_seq[i] == "TAG":
            list_STOP.append(i)
        if orf_seq[i] == "TGA":
            list_STOP.append(i)
        if orf_seq[i] == "TAA":
            list_STOP.append(i)
    return list_STOP
```

```python
def extract_orfs(orf_seq):
    START = find_START(orf_seq)
    STOP = find_STOP(orf_seq)
    orfs = []

    for a in START:
        # find the first stop codon that comes after this start
        next_stop = None
        for b in STOP:
            if b > a:
                next_stop = b
                break
        
        # if no stop codon occurs after this start, skip it
        if next_stop is None:
            continue
        
        # extract codons from start to just before stop
        orf = orf_seq[a:next_stop]
        orfs.append(orf)
    
    return orfs
```

5. How to transcribe those codons into rna.

```python
list_orfs = []
for item in orfs:
    list_orfs.append("".join(item))
```

```python
def translate_DNA(list_orfs):
    list_orfs_rna = []
    for string in list_orfs:
        list_orfs_rna.append(string.replace("T", "U"))
    return list_orfs_rna
```

```python
def codon_list(s):
    step = len(s)/3
    s_codon = []
    for codon in range(int(step)):
        start = codon * 3
        end = start + 3
        s_codon.append(s[start:end])
    return s_codon
```

```python
# codon_list(rna_list[0])
```

```python
def translate_codons(s_codon):
    result= ""
    for codons in s_codon:
        result = result + util.codon_dict[codons]
    return result
```

```python
# translate_codons(codon_list(rna_list[0]))
```

```python
def translate_pro (rna_list):
    pro_list = []
    for rna in rna_list:
        pro_list.append(translate_codons(codon_list(rna)))
    return pro_list
```

```python
def get_all_orfs(pos0, pos1, pos2, revc_pos0, revc_pos1, revc_pos2):
    all_orfs = []
    # forward strand
    for frame in [pos0, pos1, pos2]:
        orfs_in_frame = extract_orfs(frame)
        for orf in orfs_in_frame:
            list_orfs = []
            list_orfs.append("".join(orf))
            rna_list = translate_DNA(list_orfs)
            pro_list = translate_pro(rna_list)
            all_orfs.extend(pro_list)
    # reverse strand
    for rev_frame in [revc_pos0, revc_pos1, revc_pos2]:
        orfs_in_rev_frame = extract_orfs(rev_frame)
        for orf in orfs_in_rev_frame:
            list_rev_orfs = []
            list_rev_orfs.append("".join(orf))
            rna_list = translate_DNA(list_rev_orfs)
            pro_list = translate_pro(rna_list)
            all_orfs.extend(pro_list)
    return all_orfs
```
