```python
def rev_comp(DNA):
    rev_DNA = ""
    for char in DNA:
        rev_DNA = char + rev_DNA
        rev_dna = rev_DNA.casefold()
    rev_comp_DNA = ""
    for base in rev_dna:
        if base == "t":
            rev_comp_DNA = rev_comp_DNA + "A" 
        elif base == "g":
            rev_comp_DNA = rev_comp_DNA + "C"
        elif base == "c":
            rev_comp_DNA = rev_comp_DNA + "G"
        else: 
            rev_comp_DNA = rev_comp_DNA + "T"

    return rev_comp_DNA
```
