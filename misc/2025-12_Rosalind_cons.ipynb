```python
import util
```

```python
# --------------------------------------------------
# Load data
# --------------------------------------------------
lines = util.read_file_into_lines("/home/uschi/Python_homework_stefan/misc/rosalind_cons.txt")
sequences = util.parse_fasta(lines)
```

```python
# --------------------------------------------------
# Count bases
# --------------------------------------------------

def count_base(x):
    if x == 'A':
        return (1, 0, 0, 0)
    elif x == 'C':
        return (0, 1, 0, 0)
    elif x == 'G':
        return (0, 0, 1, 0)
    elif x == 'T':
        return (0, 0, 0, 1)
    else:
        return (0, 0, 0, 0)
```

```python
# --------------------------------------------------
# Count bases at one position
# --------------------------------------------------


def count_bases_in_position(sequences, position):
    base_counts = [0, 0, 0, 0] # A, C, G, T

    for seq in sequences.values():
        a, c, g, t = count_base(seq[position])
        base_counts[0] = base_counts[0] + a
        base_counts[1] = base_counts[1] + c
        base_counts[2] = base_counts[2] + g
        base_counts[3] = base_counts[3] + t

    return base_counts
```

```python
# --------------------------------------------------
# Build profile matrix
# --------------------------------------------------


# length of DNA strings
for first_key in sequences:
    string_length = len(sequences[first_key])
    break

profile_matrix = []

pos = 0
while pos < string_length:
    counts = count_bases_in_position(sequences, pos)
    profile_matrix.append(counts)
    pos += 1
```

```python
print(len(profile_matrix))
```

```python
# --------------------------------------------------
# Build consensus sequences (with branching)
# --------------------------------------------------


index_to_base = ['A', 'C', 'G', 'T']


# start with one empty sequence
consensus_list = [""]


col_index = 0
while col_index < len(profile_matrix):
    column = profile_matrix[col_index]
    
    
    # find maximum count
    max_count = column[0]
    i = 1
    while i < 4:
        if column[i] > max_count:
            max_count = column[i]
        i += 1

    # find all indices with that max count
    max_indices = []
    i = 0
    while i < 4:
        if column[i] == max_count:
            max_indices.append(i)
        i += 1

    # build new consensus list
    new_consensus = []
    
    
    for seq in consensus_list:
        for idx in max_indices:
            new_seq = seq + index_to_base[idx]
            new_consensus.append(new_seq)

    consensus_list = new_consensus
    col_index += 1
```

```python
# --------------------------------------------------
# Output
# --------------------------------------------------


print("Consensus sequences:")
for c in consensus_list:
    print(c)


print("\nProfile matrix:")


# print A row
print("A:", end=" ")
i = 0
while i < len(profile_matrix):
    print(profile_matrix[i][0], end=" ")
    i += 1
print()


# print C row
print("C:", end=" ")
i = 0
while i < len(profile_matrix):
    print(profile_matrix[i][1], end=" ")
    i += 1
print()


# print G row
print("G:", end=" ")
i = 0
while i < len(profile_matrix):
    print(profile_matrix[i][2], end=" ")
    i += 1
print()


# print T row
print("T:", end=" ")
i = 0
while i < len(profile_matrix):
    print(profile_matrix[i][3], end=" ")
    i += 1
print()
```
