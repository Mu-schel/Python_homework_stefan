Rosalind Cons

------

finding consensus sequence from multiple seqences

given at most 10 sequences of equal length with no more then 1 kb in FASTA format

return consensus sequence(s) and profile matrix

```python
import util
```

```python
lines = util.read_file_into_lines("/home/uschi/Python_homework_stefan/misc/rosalind_cons")
sequences = util.parse_fasta(lines)
```

```python
# len(sequences.get("Rosalind_1"))
sequences
```

```python
def count_base (x):
    count_A = 0
    count_C = 0
    count_G = 0
    count_T = 0
    
    if x == 'A':
        count_A += 1
    elif x == 'C':
        count_C += 1
    elif x == 'G':
        count_G += 1
    elif x == 'T':
        count_T += 1
    
    return count_A, count_C, count_G, count_T
```

```python
string_length = len(sequences.get("Rosalind_1"))

for i in range(string_length):
    base_counts = [0, 0, 0, 0]  # A, C, G, T
    for seq in sequences.values():
        a, c, g, t = count_base(seq[i])
        base_counts[0] += a
        base_counts[1] += c
        base_counts[2] += g
        base_counts[3] += t
    print(f"Position {i+1}: A:{base_counts[0]} C:{base_counts[1]} G:{base_counts[2]} T:{base_counts[3]}")
```
