```python
def count_base (x):
    count_A = 0
    count_C = 0
    count_G = 0
    count_T = 0
    
    if x == 'A':
        count_A += 1
    elif x == 'C':
        count_C += 1
    elif x == 'G':
        count_G += 1
    elif x == 'T':
        count_T += 1
    
    return count_A, count_C, count_G, count_T
```

```python
def count_bases_in_position (sequences, position):
    base_counts = [0, 0, 0, 0]  # A, C, G, T
    for seq in sequences.values():
        a, c, g, t = count_base(seq[position])
        base_counts[0] += a
        base_counts[1] += c
        base_counts[2] += g
        base_counts[3] += t
    return base_counts

# string_length = len(sequences.get("Rosalind_1"))
# # works only if all sequences have the same length
# 
# for i in range(string_length):
#     base_counts = [0, 0, 0, 0]  # A, C, G, T
#     for seq in sequences.values():
#         a, c, g, t = count_base(seq[i])
#         base_counts[0] += a
#         base_counts[1] += c
#         base_counts[2] += g
#         base_counts[3] += t
#     print(f"Position {i+1}: A:{base_counts[0]} C:{base_counts[1]} G:{base_counts[2]} T:{base_counts[3]}")
```

```python
def find_max_index_and_position (sub_list):
    max_count = max(sub_list)
    max_index = sub_list.index(max_count)
    if sub_list.count (max_count) > 1:
        all_indexes = [i for i, x in enumerate(sub_list) if x == max_count]
        return all_indexes, max_count
    return max_index, max_count
```
