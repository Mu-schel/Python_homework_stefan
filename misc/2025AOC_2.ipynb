```python
def make_full_range_list(pair):
    start = int(pair[0])
    stop = int(pair[1]) + 1
    full_range = []

    # Build the range list manually
    for n in range(start, stop):
        full_range.append(n)

    return full_range
```

```python
def get_all_full_range_lists(range_pairs):
    all_ranges = []

    for pair in range_pairs:
        full_list = make_full_range_list(pair)
        all_ranges.append(full_list)
        
    return all_ranges
```

```python
def is_invalid(s):
    if len(s) % 2 == 1:
        return False
    half_string_len = int(len(s)/2)
    first_half = s[0:half_string_len]
    second_half = s[half_string_len:]
    if first_half == second_half:
        return True
    else:
        return False
```

#### Second Part
-------------------------

- in part 1, we used first_part == second_part; but we could also think of the string as s = p + p, where p is the repeated pattern.
- here, the pattern does not have to be half of s, but it could be smaller. Still, each invalid ID will be of the form s = p + p + ... + p
- told otherwise, each invalid ID will be of the form s = p * k, where k is some integer
- what could the value of k be? How does it relate to the length of p and s?
- p is of course len(s) // len(p)
- since s = p + p + ... + p, if we know the length of p, it means that the first len(p) letters of s are the pattern!
- we don't know len(p) a priori, but we know how long p could be at most
- namely, len(s) // 2 (because this is the longest subword that, if repeated, will create s)
- so given a word s, we can try out all the different sub-words p that could make up s if repeated multiple times...


```python
def more_invalid(STR):
    for length_p in range(1, (len(STR)//2)+1):
        p = STR[0:length_p]
        k = len(STR) // length_p
        fake_str = p * k
        if fake_str == STR:
            return STR
    # return None
```

```python
all_invalid_ids = []
for a in all_ranges:
    for number in a:
        #print(number)
        check = more_invalid(str(number))
        #print(check)
        if check is not None:
            all_invalid_ids.append(check)
```
