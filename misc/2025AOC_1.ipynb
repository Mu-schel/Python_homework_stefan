```python
# works if there are only up to three digits in the 

def find_pos (List_of_moves , start_position):
    pos = start_position
    position_result = []

    for item in List_of_moves:
        if item[0] == 'R' and int(item[1:]) >= 100:
            pos += (int(item[1:])-((int(item[1])) * 100))
            if pos >= 100:
                pos -= 100
            position_result.append(pos)
        elif item[0] == 'L' and int(item[1:]) >= 100:
            pos -= (int(item[1:])-((int(item[1])) * 100))
            if pos < 0:
                pos += 100
            position_result.append(pos)
        elif item[0] == 'R' and int(item[1:]) < 100:
            pos += int(item[1:])
            if pos >= 100:
                pos -= 100
            position_result.append(pos)
        elif item[0] == 'L' and int(item[1:]) < 100:
            pos -= int(item[1:])
            if pos < 0:
                pos += 100
            position_result.append(pos)
```
