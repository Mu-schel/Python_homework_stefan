```python
List_test= [
"L368",
"L530",
"R48",
"L5",
"R60",
"L55",
"L1",
"L99",
"R14",
"L382",
]
```

```python
pos = 50

for item in List_test:
    if item[0] == 'R' and int(item[1:]) >= 100:
        pos += (int(item[1:])-((int(item[1])) * 100))
        if pos >= 100:
            pos -= 100
        print(pos , "R1")
    elif item[0] == 'L' and int(item[1:]) >= 100:
        pos -= (int(item[1:])-((int(item[1])) * 100))
        if pos < 0:
            pos += 100
        print(pos, "L1")
    elif item[0] == 'R' and int(item[1:]) < 100:
        pos += int(item[1:])
        if pos >= 100:
            pos -= 100
        print(pos , "R2")
    elif item[0] == 'L' and int(item[1:]) < 100:
        pos -= int(item[1:])
        if pos < 0:
            pos += 100
        print(pos , "L2")
```
