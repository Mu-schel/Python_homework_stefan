#### Calculating GC content of a given DNA sequence

```python
dna = "CCACCCTCGTGGTATGGCTAGGCATTCAGGAACCGGAGAACGCTTCAGACCAGCCCGGACTGGGAACCTGCGGGCAGTAGGTGGAAT"
count_C = dna.count("C")
count_G = dna.count("G")
G_C_count = ((count_C + count_G)*100)/ len(dna)
G_C_count = round(G_C_count , 1)
print("The sequence contains" , count_C , "C and" , count_G , "G and has a GC content of" , G_C_count , "percent.")
```
