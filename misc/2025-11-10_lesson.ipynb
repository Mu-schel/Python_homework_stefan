Making nested dictionaries.

```python
grades = {'Juan': {'Maths': 2},
          'John': {'History': 1, 'Biology': 3},
          'Gianni': {'Biology': 1}
         }
grades['Juan']["History"] = 2
grades['Juan']["iology"] = 1

grades["John"]['Maths'] = 1

grades['Gianni']['History'] = 4
grades['Gianni']["Maths"] = 3

print(grades)
```

### Loops

```python
# for <vaiable> in <collection:
# do something (this part is the <body>

for number in [1, 2, 3, 4, 5]:
    print(number)
```

```python
s = "the quick brown fox jumps over the lazy dog"
```

```python
for characters in s:
    print(characters)
```

```python
# range for doing something x times in the range
for i in range(14):
    print(i, "Hello world.")
```

```python
# better then doing the range is letting enumerate count the items in the collection
for i, item in enumerate(s.split()):
    print(i, item)
```

```python
#chellenges in the carpentries course

total = 0
for char in "tin":
    total = total + 1
    print(char, total)
```

```python
original = "fish"
result = ""
for char in original:
    result = char + result
print(result)
```

```python
total = 0
for word in ["red", "green", "blue"]:
    total = total + len(word)
print(total)
```

```python
lengths = []
for word in ["red", "green", "blue"]:
    lengths.append(len(word))
print(lengths)
```
