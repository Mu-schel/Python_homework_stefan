```python
def range_overlap(a, b):
    i = int(a[0])
    j = int(a[1])
    l = int(b[0])
    k = int(b[1])
    if i<=l<=j<=k:
        return True
    if l<=i<=j<=k:
        return True
    if i<=l<=k<=j:
        return True
    if l<=i<=k<=j:
        return True
    if i==k and l<j:
        return True
    if i==k and j<l:
        return True
    if j==l and i<k:
        return True    
    if j==l and k<l:
        return True
    return False
```

```python
def merge_ranges(a, b):
    i = int(a[0])
    j = int(a[1])
    l = int(b[0])
    k = int(b[1])
    new_start = min(i,l)
    new_end = max(j,k)
    return [str(new_start), str(new_end)]
```
