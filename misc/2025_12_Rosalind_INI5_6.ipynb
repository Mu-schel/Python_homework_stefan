Ini 6 dictionaries

get a strin with at most 10k words 

return a dictionary with all the words and their count

while being case sensetive

```python
test_string_list = util.read_file_into_lines("/home/uschi/Python_homework_stefan/misc/rosalind_ini6.txt")
```

```python
test_string = test_string_list[0]
word_list = test_string.split(' ')
# word_list
```

```python
# make a dictionary with each word as key, so we can count in the next step
word_dict = {}
for word in word_list:
    if word in word_dict:
        word_dict[word] += 1
    else:
        word_dict[word] = 1
```

```python
# word_dict
```

```python
for key, value in word_dict.items():
    print(key,"",value)
```
